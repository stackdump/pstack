```python
import psycopg2 as psycopg

# Connect to your postgres DB
conn = psycopg.connect("dbname=pstack user=pstack password=development")

# Open a cursor to perform database operations
cur = conn.cursor()

def get_rates():
     cur.execute("SELECT get_rates()")
     return cur.fetchone()[0]

def insert_one():
     cur.execute("insert into blocks  select max(ht)-1 as ht from blocks;")
     return cur.rowcount
print(insert_one())
conn.commit()
```
